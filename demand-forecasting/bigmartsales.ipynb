### 4. 컬럼들 한국어로 바꾸기

***컬럼이 object인 것만 추출하기***

```python
text_cols = df.select_dtypes(include="object").columns
print(text_cols)
```

***Item_Fat_Content 번역***

```python
df[["Item_Fat_Content"]]
item_fat_content_map = df["Item_Fat_Content"].value_counts()

# Item_Fat_Content
item_fat_content_translation_dict = {
    "Low Fat": "저지방",
    "low fat": "저지방",
    "LF": "저지방",
    "Regular": "일반",
    "reg": "일반"
}

item_fat_content_translated = item_fat_content_map.rename(index=item_fat_content_translation_dict)

print(item_fat_content_translated)
```

***Item_Type 번역***

```python
df[["Item_Type"]]

# Item_Type의 값 개수 확인
item_type_map = df["Item_Type"].value_counts()

# Item_Type
item_type_translation_dict = {
    "Fruits and Vegetables": "과일과 채소",
    "Snack Foods": "간식",
    "Household": "생활용품",
    "Frozen Foods": "냉동식품",
    "Dairy": "유제품",
    "Canned": "통조림",
    "Baking Goods": "베이킹 재료",
    "Health and Hygiene": "건강/위생",
    "Soft Drinks": "탄산음료",
    "Meat": "육류",
    "Breads": "빵",
    "Hard Drinks": "주류",
    "Others": "기타",
    "Starchy Foods": "전분&녹말 식품",
    "Breakfast": "조식",
    "Seafood": "해산물"
}

item_type_map_translated = item_type_map.rename(index=item_type_translation_dict)

print(item_type_map_translated)
```
